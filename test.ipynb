code to generate create table and insert table syntax


```python
import csv

# Read the CSV file to extract the district names from the first row
file_path = 'District_average_daily1.csv'
with open(file_path, newline='') as csvfile:
    csvreader = csv.reader(csvfile)
    headers = next(csvreader)  # Get the first row

# Prepare the CREATE TABLE statement
table_name = 'District_average_daily1'
create_table_sql = f"CREATE TABLE {table_name} (\n"
create_table_sql += "  `Dates` DATE NOT NULL,\n"

# Add columns for each district (skip the first header since it's 'date')
for district in headers[1:]:
    create_table_sql += f"  `{district}` FLOAT,\n"
create_table_sql = create_table_sql.rstrip(',\n') + "\n);"

# Prepare the LOAD DATA INFILE statement
load_data_sql = f"""
LOAD DATA INFILE '{file_path}'
INTO TABLE {table_name}
FIELDS TERMINATED BY ','
ENCLOSED BY '"'
LINES TERMINATED BY '\\n'
IGNORE 1 ROWS
(`Dates`, {', '.join([f'`{district}`' for district in headers[1:]])});
"""

# Print the SQL statements
# print(create_table_sql)
print(load_data_sql)





```

Code to change date format


```python
import csv
from datetime import datetime

file_path = 'District_average_daily1.csv'

# Function to convert date format
def convert_date_format(date_str):
    # Assuming date_str is in dd-mm-yyyy format
    date_obj = datetime.strptime(date_str, '%d-%m-%Y')
    return date_obj.strftime('%Y-%m-%d')


updated_data = []

with open(file_path, newline='') as csvfile:
    csvreader = csv.reader(csvfile)
    for i, row in enumerate(csvreader):
        if i == 0:
            # Append headers as-is
            updated_data.append(row)
        else:
            # Convert date format for column 1 (index 0)
            row[0] = convert_date_format(row[0])
            updated_data.append(row)

# Write the updated data back to the CSV file
with open(file_path, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerows(updated_data)

print(f"Date format conversion completed for {len(updated_data) - 1} rows.")
```
